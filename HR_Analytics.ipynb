```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

print("Setup Ready")
```

```python
df = pd.read_csv("../data/hr_data.csv")

df.head()
```

```python
df.shape
```

```python
salary_department = (
    df.groupby('Department')['MonthlyIncome']
    .mean()
    .sort_values(ascending=False)
)

salary_department
```

```python
salary_department = (
    df.groupby('Department')['MonthlyIncome']
    .mean()
    .sort_values(ascending=False)
)

salary_department
```

```python
plt.figure(figsize=(8,5))

salary_department.plot(kind='bar')

plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Monthly Income")

plt.xticks(rotation=45)
plt.show()
```

```python
job_roles = df['JobRole'].value_counts()

job_roles
```

```python
plt.figure(figsize=(10,5))

job_roles.plot(kind='bar')

plt.title("Employees by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Number of Employees")

plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x='Age',
    bins=20
)

plt.title("Employee Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Employees")

plt.show()
```

```python
overtime_attrition = pd.crosstab(
    df['OverTime'],
    df['Attrition']
)

overtime_attrition
```

```python
print("""
HR Analytics Key Insights:

1. Employee attrition was analysed to understand workforce turnover.
2. Department and job role analysis identified workforce distribution patterns.
3. Salary analysis showed differences in average income across departments.
4. Overtime analysis helped identify potential employee retention risks.
5. Age distribution provided insights into employee demographics.
""")
```
